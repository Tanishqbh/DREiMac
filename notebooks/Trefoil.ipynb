### Trefoil knot

The purpose of this notebook is to demonstrate the use of the <tt>perc</tt> parameter, which tells the circular coordinates algorithm at which point of the filtration the circle-valued map must be constructed.


```python
from dreimac import CircularCoords, GeometryExamples, PlotUtils
from persim import plot_diagrams
```

The data consists of a trefoil knot in 3D. As such, the data, at a small scale, is parametrized by a circle looping around twice. At a larger scale, though, the data looks like a single circle. We will identify this behavior by finding two circle-valued maps using the same persistent cohomology class, but two values of the <tt> perc </tt> parameter.

```python
X = GeometryExamples.trefoil(n_samples = 2500, horizontal_width=10)
PlotUtils.plot_2d_scatter_with_different_colorings(X)
```

The persistence diagram clearly shows a single high prominence class.

```python
cc = CircularCoords(X, 300, prime=3)
plot_diagrams(cc.dgms_)
```

We now choose a small <tt>perc</tt> parameter and get a circle-valued representation that parametrized the small-scale circularity of the data.

```python
perc = 0.1
circular_coords = cc.get_coordinates(perc=perc, cocycle_idx=0)

PlotUtils.plot_2d_scatter_with_different_colorings(
    X,
    [circular_coords],
    point_size=10,
    cmap="hsv",
)
```

Choosing a large <tt>perc</tt>, we get the following.

```python
large_perc = 0.8
circular_coords_large_perc = cc.get_coordinates(perc=large_perc, cocycle_idx=0)
```

```python
PlotUtils.plot_2d_scatter_with_different_colorings(
    X,
    [circular_coords_large_perc],
    point_size=10,
    cmap="hsv",
)
```
